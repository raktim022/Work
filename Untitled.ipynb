```python
from transformers import pipeline

nlp = pipeline(
    "document-question-answering",
    model="impira/layoutlm-document-qa",
)
```

```python
import cv2
import matplotlib.pyplot as plt
```

```python
ques=["What is the name of the store?","What is the address of the store?","What is the total amount?"]
answers=["Name of store","Address of store","Total amount"]
out=[]
for i in range(1,7):
    ans=""
    image_path=f"images/{i}.jpg"
    for q in range(len(ques)):
        res=nlp(image_path,ques[q])
        #print(res[0]["answer"])
        ans+=answers[q]+":"+res[0]["answer"]+"  "
    out.append(ans)
```

```python
plt.imshow(cv2.imread("images/1.jpg"))
print(out[0])
```
